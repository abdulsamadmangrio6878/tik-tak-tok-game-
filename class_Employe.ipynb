```python
class Employeee:
    def __init__(self, first_name, last_name,ssn):
        self.first_name = first_name
        self.last_name = last_name
        self.ssn = ssn

    def get_first_name(self):
        return self.first_name

    def get_last_name(self):
        return self.last_name

    def get_social_security_no(self):
        return self.ssn

    def __str__(self) -> str:
        return f"NAME: {self.first_name} {self.last_name}\nSOCISAL SECURITY NO: {self.ssn}"

class CommissionEmployeee(Employeee):
    def __init__(self, first_name, last_name, ssn,comission_rate,gross_sales):
        super().__init__(first_name, last_name, ssn)
        self.comission = comission_rate
        self.gross = gross_sales


    def earnings(self):
        return self.comission*self.gross
        


    def __str__(self) -> str:
        return super().__str__() + f'\n{self.comission}\nGross sales,{self.gross}'


class HourlyEmployeee(Employeee):
    def __init__(self, first_name, last_name, ssn,hourly_wage, hours_worked):
        super().__init__(first_name, last_name, ssn)
        self.set_wage(hourly_wage)
        self.set_hours(hours_worked)

    def set_wage(self,hourly_wage):
        if hourly_wage < 0:
            raise ValueError("Wage connot be negative.")
        self._wage = hourly_wage

    def get_wage(self):
        return self._wage

    def set_hours(self,hours_worked):
        if hours_worked < 0 or hours_worked > 168:
            raise ValueError("Hours worked must be between 0 and 168.")

        self._hours = hours_worked
    
    def get_hours(self):
        return self._hours


    def earnings(self):
        if self._hours <= 40:
            return self._wage + (self._hours - 40)*self._wage * 1.5

    def __str__(self) -> str:
        return super().__str__() + f"\nWAGE: {self._wage}\nHours worked: {self._hours}\nEarnings: {self.earnings()}"


hee = HourlyEmployeee('John','Doe','124-45-6633',10,50)

print(hee.get_first_name())
print(hee.get_last_name())
print(hee.get_wage())
print(hee.get_hours())
print(hee.earnings())
print(hee)


cee = CommissionEmployeee('John','Doe','124-45-6633',0.10, 10000)


print(cee.get_first_name())
print(cee.get_first_name())
print(cee.get_social_security_no())
print(cee.earnings())
print(cee)
```
